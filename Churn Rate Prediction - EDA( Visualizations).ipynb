#### 1. Total Number of Bookings

#### 2. Top 10 Countries of Origin of Guests

```python
data.country.value_counts().head(10)
```

```python
d1 = pd.value_counts(data['country']).iloc[:10].index #shows index (name) of top 10 countries
d1
```

```python
fig, ax= plt.subplots(figsize=(15,8))
sns.countplot(x = 'country', data = data, order = pd.value_counts(data['country']).iloc[:10].index, palette = 'colorblind', ax=ax)
plt.title('Top 10 Countries by the Origin of the Guests', weight ='bold')
plt.xlabel('Country')
plt.ylabel('Reservation Count')
```

The above graph shows that the majority of bookings were done from Portugal (PRT) about 40% of the total bookings, followed by Great Britain (GBR) at 10%.

#### 3. Which Hotel had more cancellations

```python
data.groupby(['is_canceled'])['hotel'].value_counts()
```

```python
fig, ax= plt.subplots(figsize=(15,8))
sns.set_style('whitegrid')
sns.countplot(x='hotel', hue = 'is_canceled', data=data, ax=ax, palette = 'deep')
plt.title('Number of Cancellations by Hotel Type', weight ='bold')
plt.xlabel('Hotel', weight = 'bold')
plt.ylabel('Cancellation Count', weight = 'bold')
```

City Hotels had more cancellations than Resort Hotels (about 75%).

#### 4. Percentage of Cancellations Vs Non-Cancellations

```python
percentage = data.is_canceled.value_counts(normalize=True)
labels = ['Not Cancelled', 'Cancelled']
plt.title('Percentage of Cancelled Vs. Not Cancelled Bookings', weight = 'bold')
plt.pie(percentage, labels=labels, autopct = '%1.1f%%', wedgeprops={'edgecolor':'black'})
plt.show()
```

37% of the total bookings were cancelled according to the pie chart.

#### 5. Busiest Month at the Hotels

```python
data['arrival_date_month'].value_counts()
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
data['arrival_date_month'] = pd.Categorical(data['arrival_date_month'], categories = months, ordered=True)
fig, ax= plt.subplots(figsize=(12,5))
sns.countplot(x='arrival_date_month', data = data,palette= 'Set1', ax=ax)
plt.title('Busiest Months at the Hotel', weight='bold')
plt.xlabel('Month')
plt.xticks(rotation=30)
plt.ylabel('Reservation Count')
```

* As seen from the graph above, August is the busiest month of the year followed by July. 
* January appears to be least occupied month of the year.

#### 6. Percentage of Cancellations per Month for each Hotel

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
data['arrival_date_month'] = pd.Categorical(data['arrival_date_month'], categories = months, ordered=True)
fig, ax= plt.subplots(figsize=(12,5))
sns.barplot(x='arrival_date_month', y='is_canceled',hue='hotel',data = data, palette= 'Set2', ax=ax)
plt.title('Percentage of Cancellations per month for each Hotel', weight='bold')
plt.xlabel('Month')
plt.xticks(rotation=30)
plt.ylabel('Cancellation Percentage')
plt.show()
```

* For City hotels, there are about 40% cancellations every month of the year.
* For Resort hotels, the cancellations seem to vary, reaching the highest values in the month of June, August, September. During winters, it can be observed that the cancellations are at the lowest! That is not the case for City Hotel.

### 8. Year of Arrival Vs Lead Time for Each Cancellation

```python
data.arrival_date_year.value_counts()
```

```python
fig, ax = plt.subplots(figsize=(12,5))
sns.barplot(data=data,x = 'arrival_date_year', y='lead_time', hue = 'is_canceled')
plt.title('Year of Arrival Vs Lead for Cancellations', weight='bold')
plt.xlabel('Year')
plt.ylabel('Lead Time')
```

* Lead Time is defined as the difference between the booking date and the actual arrival date.
* Cancellation appears to increase as the lead time exceeds 100 days.
* 2015 had the most cancellations because of the highest lead time of about 160 days as compared to 2016 and 2017.

### 9. Average Daily Rate(adr) Vs Month of Arrival by Booking cancellations

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
data['arrival_date_month'] = pd.Categorical(data['arrival_date_month'], categories = months, ordered=True)
fig, ax= plt.subplots(figsize=(12,5))
sns.lineplot(x='arrival_date_month', y='adr',hue='hotel',data = data, palette= 'Set1', ax=ax)
plt.title('Average Daily Rate Vs Month of Arrival for Hotels', weight='bold')
plt.xlabel('Month')
plt.xticks(rotation=30)
plt.ylabel('Average Daily Rate')
plt.show()
```

```python
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
data['arrival_date_month'] = pd.Categorical(data['arrival_date_month'], categories = months, ordered=True)
fig, ax= plt.subplots(figsize=(12,5))
sns.lineplot(x='arrival_date_month', y='adr',hue='is_canceled',data = data, palette= 'Set1', ax=ax)
plt.title('Average Daily Rate Vs Month of Arrival for Hotels', weight='bold')
plt.xlabel('Month')
plt.xticks(rotation=30)
plt.ylabel('Average Daily Rate')
plt.show()
```

* The average daily rate for city hotel is more in the months of April, May, June and then tends to go down gradually.
* The average daily rate for resort hotel appears to exponentially increase from June till August which are also the busiest months.
* When the ADR is at its peak in the month of August, it appears that the number of cancellations are also at the all-time high!
